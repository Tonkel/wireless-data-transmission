```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import zlib
```

```python
depth =  np.uint8(np.random.randint(255,size = (480,640)))
```

```python
def Depth2Gray(d4d):
        d4d = 255 - cv2.cvtColor(d4d, cv2.COLOR_GRAY2RGB)
        return d4d
```

```python
Depth2Gray(depth).shape
```

```python
d4d = Depth2Gray(depth)
plt.imshow(d4d)
plt.show()
print("Origin Depth Array Data Size: "+str(len(d4d.tostring())))
```

```python
print("After Native Zlib Compression:", len(zlib.compress(d4d.tostring())), "; Compression Rate:",len(zlib.compress(d4d.tostring()))*1.0/len(d4d.tostring()))
```

```python
color = np.uint8(np.random.randint(255,size = (480,640,3)))
```

```python
plt.imshow(color)
plt.show()
print("Origin color Array Data Size: "+str(len(color.tostring())))
```

```python
print("After Native Zlib Compression:", len(zlib.compress(d4d.tostring())), "; Compression Rate:",len(zlib.compress(d4d.tostring()))*1.0/len(d4d.tostring()))
```

```python
plt.imshow("Depth | Color", np.hstack((d4d,color)))
plt.show()
```

```python
rgbd = np.dstack((color, depth))
```

```python
print("Before Compression:",
      len(rgbd.tostring()),
      "After Native Zlib Compression:", len(zlib.compress(rgbd.tostring())), "; Compression Rate:",len(zlib.compress(rgbd.tostring()))*1.0/len(rgbd.tostring()))
```
